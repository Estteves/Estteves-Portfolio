# **Análise Exploratória de Dados de Logística**

## 1\. Contexto


O Loggi Benchmark for Urban Delivery contém conjuntos de dados e scripts de benchmark para problemas de grande escala usando como teste algumas das maiores cidades do Brasil. O conjunto de dados simula o desafio de uma grande empresa de entrega na última milha da cadeia de suprimentos. Em todo o mundo, áreas metropolitanas como o Rio de Janeiro testemunham a entrega de dezenas ou mesmo centenas de milhares de encomendas e correspondências. As entregas nessas áreas são inesperadamente distribuídas de forma muito desigual. Áreas ricas e densamente povoadas conduzem a maior parte das entregas. Além dos fatores econômicos, os dados também consideram a geografia dos locais representados e as distâncias reais de deslocamento nas ruas.
O objetivo é realizar uma análise exploratória dos dados utilizando as bibliotecas Python json e pandas.

## 2\. Pacotes e bibliotecas

```
# importe todas as suas bibliotecas aqui, siga os padrões do PEP8:
import json
import pandas as pd
```

## 3\. Exploração de dados

```
# faça o código de exploração de dados:
#Coletar e carregar dados
!wget -q "https://raw.githubusercontent.com/andre-marcos-perez/ebac-course-utils/main/dataset/deliveries.json" -O deliveries.json

with open('deliveries.json', mode='r', encoding='utf8') as file:
  data = json.load(file)
```

```
# - wrangling da estrutura;
deliveries_df = pd.DataFrame(data)
deliveries_df.head()
```

a coluna origin contem dados nested ou aninhados na estrutura do JSON. Vamos normalizar a coluna com uma operação conhecida como flatten ou achatamento que transforma cada chave do JSON em uma nova coluna.

```
hub_origin_df = pd.json_normalize(deliveries_df["origin"])
hub_origin_df.head()
```

Com o dados achatados, vamos junta-los ao conjunto de dados principal:

```
deliveries_df = pd.merge(left=deliveries_df, right=hub_origin_df, how='inner', left_index=True, right_index=True)
deliveries_df.head()
```

Após separar os dados da coluna 'origin' vamos removê-la do Dataframe 'deliveries_df' e alterar a posição das colunas 'lng' e 'lat'

```
deliveries_df = deliveries_df.drop("origin", axis=1)
deliveries_df = deliveries_df[["name", "region", "lng", "lat", "vehicle_capacity", "deliveries"]]
deliveries_df.head()
```

Agora, vamos renomear as colunas 'lng' e 'lat'

```
deliveries_df.rename(columns={"lng": "hub_lng", "lat": "hub_lat"}, inplace=True)
deliveries_df.head()
```

A coluna deliveries contem dados uma lista de dados nested ou aninhados na estrutura do JSON. Vamos normalizar a coluna com uma operação conhecida como explode ou explosão que transforma cada elemento da lista em uma linha. Por fim, faremos os flatten ou achatamento do resultado coluna:

```
deliveries_exploded_df = deliveries_df[["deliveries"]].explode("deliveries")
deliveries_exploded_df.head()
```

```
deliveries_normalized_df = pd.concat([
  pd.DataFrame(deliveries_exploded_df["deliveries"].apply(lambda record: record["size"])).rename(columns={"deliveries": "delivery_size"}),
  pd.DataFrame(deliveries_exploded_df["deliveries"].apply(lambda record: record["point"]["lng"])).rename(columns={"deliveries": "delivery_lng"}),
  pd.DataFrame(deliveries_exploded_df["deliveries"].apply(lambda record: record["point"]["lat"])).rename(columns={"deliveries": "delivery_lat"}),
], axis= 1)
deliveries_normalized_df.head()
```

Após separar os dados da coluna 'deliveries' vamos removê-la do Dataframe 'deliveries_df' e concatenar a direita os dados do 'deliveries_normalized_df' com o 'deliveries_df'

```
deliveries_df = deliveries_df.drop("deliveries", axis=1)
deliveries_df = pd.merge(left=deliveries_df, right=deliveries_normalized_df, how='right', left_index=True, right_index=True)
deliveries_df.reset_index(inplace=True, drop=True)
deliveries_df.head()
```
